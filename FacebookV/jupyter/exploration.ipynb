```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../input/train09_100.csv')
```

```python
data.head()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(data['x'], data['y'], s=1, c=data['place_id'], linewidths=0)
plt.show()
```

```python
MINUTES_IN_DAY = 60 * 24
MINUTES_IN_MONTH = 30 * MINUTES_IN_DAY
START = min(data['time'])

day_of_week = ((data['time'] - START) // MINUTES_IN_DAY) % 7
time_of_day = (data['time'] - START) % MINUTES_IN_DAY

data['day_of_week'] = day_of_week
data['time_of_day'] = time_of_day
```

```python
data.head()
```

```python
M = 3
for i in range(M):
    place_id = data['place_id'].unique()[i]
    plt.hist(data[data['place_id']==place_id]['time_of_day'], bins='fd')
    plt.show()
```

```python
M = 3
for i in range(M):
    place_id = data['place_id'].unique()[i]
    plt.hist(data[data['place_id']==place_id]['day_of_week'], bins=7)
    plt.show()
```

```python
M = 3
for i in range(M):
    place_id = data['place_id'].unique()[10+i]
    plt.hist(data[data['place_id']==place_id]['accuracy'], bins=10)
    plt.show()
```

```python
M = 3
for i in range(M):
    place_id = data['place_id'].unique()[50+i]
    plt.hist(data[data['place_id']==place_id]['time'])
    plt.show()
```
