```python
import numpy as np
import pandas as pd
import math

train = pd.read_csv('input/train.csv')
test = pd.read_csv('input/test.csv')
```

```python
train2 = train[['Survived']]
train2['Sex'] = pd.Categorical.from_array(train['Sex']).codes + np.random.normal(0, 0.1, train2.shape[0])
train2['Pclass'] = train['Pclass'] + np.random.normal(0, 0.1, train2.shape[0])
train2['Age'] = train['Age']
train2['Age'] = train['Age'].fillna(np.mean(train['Age']))
train2['SibSp_log'] = train.apply(lambda row: math.log(row['SibSp'] + 1), axis=1) + np.random.normal(0, 0.1, train2.shape[0])
```

```python
import seaborn as sns
%matplotlib inline
```

```python
plot = sns.pairplot(train2, hue='Survived', markers=['+','o'], dropna=True)
```
